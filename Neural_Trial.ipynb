```python
import pandas as pd
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")
train.F10.unique()
codes = {' White':0, ' Black':1, ' Asian-Pac-Islander':2,  ' Amer-Indian-Eskimo':3,' Other':4}
train['F10'] = train['F10'].map(codes)
test['F10'] = test['F10'].map(codes)
codes = {' Male':0, ' Female':1}
train['F11'] = train['F11'].map(codes)
test['F11'] = test['F11'].map(codes)

train.F11.unique()
```

```python
X = (train[['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11']])
y=train[['credit']]
Y=test[['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11']]
type(X)
type(Y)
```

```python
import numpy
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
numpy.random.seed(7)
# create the model
model = Sequential()

# layers
model.add(Dense(units = 9, kernel_initializer = 'uniform', activation = 'relu' ))
model.add(Dense(units = 9, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Train the ANN
model.fit(X, y, batch_size = 32, epochs = 200)
# Final evaluation of the model
```

```python
y_pred = model.predict(Y).flatten()
print(y_pred)
ans=[]
for i in y_pred:
  if(i>=0.5):
    ans.append(1)
  else:
    ans.append(0)
print(ans)
```

```python

ans=pd.DataFrame(data=ans)
print(len(ans))
```
