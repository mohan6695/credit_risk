```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")
#one-hot encodind for categorical

pd.get_dummies(train['F10'], prefix='race')
train= pd.concat([train,pd.get_dummies(train['F10'],prefix='race')],axis=1)
train= pd.concat([train,pd.get_dummies(train['F11'],prefix='gender')],axis=1)
train.drop(['id','F10','F11'],axis=1, inplace=True)



pd.get_dummies(test['F10'], prefix='race')
test= pd.concat([test,pd.get_dummies(test['F10'],prefix='race')],axis=1)
test= pd.concat([test,pd.get_dummies(test['F11'],prefix='gender')],axis=1)
test.drop(['id','F10','F11'],axis=1, inplace=True)

test.columns
```

```python
X = (train[['F1','F2','F3','F4','F5','F6','F7','F8','F9']])
y=train[['credit']]
Y=test[['F1','F2','F3','F4','F5','F6','F7','F8','F9']]
```

```python
print((X))
```

```python
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier,ExtraTreesClassifier,AdaBoostClassifier, GradientBoostingRegressor

model= GradientBoostingClassifier(learning_rate=0.5,max_depth=5,max_features=0.3,min_samples_leaf=25,min_samples_split=4,n_estimators=100,subsample=1.0)

model.fit(X, y)
```

```python
y_pred = model.predict(Y)
y_pred=[int(round(i)) for i in y_pred]

#y_pred = [int(round(value)) for value in y_pred]
from collections import Counter
a=(Counter(y_pred))
print(a[0]+a[1])
```
